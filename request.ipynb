```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# 1. Load API key from .env file
load_dotenv()

# 2. Initialize the OpenAI client pointing to OpenRouter
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
)

print("Sending first request...")

# First API call
response = client.chat.completions.create(
    model="nex-agi/nex-n2-pro:free",
    messages=[
        {"role": "user", "content": "what quantam comupter is answer in 2 line ?"}
    ],
    
)
print(response.choices[0].message.content)


```

```python
# --- Helper Functions ---

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    # Use the OpenAI SDK format here
    response = client.chat.completions.create(
        model="nex-agi/nex-n2-pro:free", 
        messages=messages,
    )
    # Extract the text using the OpenAI format
    return response.choices[0].message.content


# --- Putting It All Together ---

# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Define quantum computing in one sentence")

# Get the model's response
answer = chat(messages)
print(f"Turn 1 Answer: {answer}\n")

# Add the model's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response with full context
final_answer = chat(messages)
print(f"Turn 2 Answer: {final_answer}")
```
